```python
import pandas as pd
import os
import re
import requests
```

```python
data_path = r'../data'
file_path = os.path.join(data_path, 'train.csv')
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
def get_critics_rating(title):
    print ('Sending request for %s' % title)
    r = requests.get(r"http://api.marcalencc.com/metacritic/movie/{0}".format(title))
    if r.status_code == 200:
        try:
            rating = r.json()[0]['Rating']
            return rating['CriticRating']/10, rating['CriticReviewCount']
        except KeyError:
            return 0,0
        
    return 0,0
```

```python
# Build a request for every title. 
# First replace .:,& with empty string. Then, replace non letters/digits with a dash
titles = df['title'].apply(lambda x: re.sub("[^\w']", "-", re.sub("(?:[:.,]|\s&)","", x)))
ratings = titles.apply(get_critics_rating)

df['critics_vote'] = [r[0] for r in ratings]
df['critics_count'] = [r[1] for r in ratings]
```

```python
df.to_csv(file_path, encoding='utf-8')
```

```python
zero_title = df[df['critics_vote'] == 0]['title'].apply(lambda x: re.sub("[^\w']", "-", re.sub("(?:[:.,]|\s&)","", x)))
zero_title_rating = zero_title.apply(get_critics_rating)
```

```python
df.loc[zero_title_rating.index, 'critics_vote'] = [r[0] for r in zero_title_rating]
df.loc[zero_title_rating.index, 'critics_count'] = [r[1] for r in zero_title_rating]
```

```python
df.to_csv(file_path, encoding='utf-8')
```
